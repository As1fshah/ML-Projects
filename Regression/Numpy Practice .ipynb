```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
student_marks = np.random.randint(24,60,size = (45,4))
```

```python
df= pd.DataFrame(student_marks,
             index = ['Student 1','Student 2','Student 3','Student 4','Student 5','Student 6','Student 7','Student 8','Student 9','Student 10',
                      'Student 11','Student 12','Student 13','Student 14','Student 15','Student 16','Student 17','Student 18','Student 19','Student 20',
                      'Student 21','Student 22','Student 23','Student 24','Student 25','Student 26','Student 27','Student 28','Student 29','Student 30',
                      'Student 31','Student 32','Student 33','Student 34','Student 35','Student 36','Student 37','Student 38','Student 39','Student 30',
                      'Student 41','Student 42','Student 43','Student 44','Student 45'],
             columns = ['Machine Learning', 'Computer Vision','Natural Language Processing','Advance Machine Learning'])
```

```python
df['Total '] = df.sum(axis = 1)
```

```python
df['Percentage'] = (df['Total '] / 240) * 100
```

```python
grading_scale = {
    (80,100) : "O",
    (70,79.99) : "A+",
    (60,69.99) : "A",
    (55,59.99) : "B+",
    (50,54.99) : "B",
    (45,49.99) : "C",
    (40,44.99) : "D",
    (0,39.99) : "F",
}
```

```python
pointer_scale = {
    ('O') : 10,
    ('A+') : 9,
    ('A') : 8,
    ('B+') : 7,
    ('B') : 6,
    ('C') : 5,
    ('D') : 4,
    ('F') : 0,
}
```

```python
def grading_sys(percentage):
  for perc, grade in grading_scale.items():
    if percentage >= perc[0] and percentage <= perc[1]:
        return grade
```

```python
cgpa_scale = {
    ('O') : 10,
    ('A+') : 9,
    ('A') : 8,
    ('B+') : 7,
    ('B') : 6,
    ('C') : 5,
    ('D') : 4,
    ('F') : 0,
    }

def cgpa_sys(cgpa):
  return cgpa_scale.get(cgpa, None)
```

```python
performance_scale = {
    (10) : 'Outstanding',
    (9) : 'Excellent',
    (8) : 'Very Good',
    (7) : 'Good',
    (6) : 'Above Average',
    (5) : 'Average',
    (4) : 'Pass',
    (0) : 'Fail',
    }

def performance_sys(cgpa):
  return performance_scale.get(cgpa)
```

```python
df['Grade'] = df['Percentage'].apply(grading_sys)
```

```python
df['Pointer'] = df['Grade'].apply(cgpa_sys)
```

```python
df['Remark'] = df['Pointer'].apply(performance_sys)
```
